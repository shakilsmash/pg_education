```python
# Library Imports
import os
import pandas as pd

from dotenv import load_dotenv
load_dotenv()
```

```python
# Loading dotenv variables
dataset_path = os.getenv('DATASET_PATH')
raw_path = os.getenv('RAW_DATASET_PATH')
test_file = os.getenv('TEST_FILE')
print(dataset_path)
print(raw_path)
print(test_file)
```

```python
# Test if all files found
all_files = os.listdir(raw_path)
total_files = len(all_files)
print(len(all_files))
```

```python
gradesheet_meta = []
count = 0
for file in all_files:    
    print("Fetching:",total_files-count,":",raw_path+"/"+file)
    count = count + 1
    try:
        data = pd.read_excel(raw_path+"/"+file,"Final GradeSheet")
    except ValueError:
        data = pd.read_excel(raw_path+"/"+file)

    try:
        gradesheet_meta.append([data['BRAC University'][1],data['BRAC University'][3],data['BRAC University'][5],data['BRAC University'][6],file])
    except KeyError:
        continue

```

```python
# Write to file
dataframe = pd.DataFrame(gradesheet_meta)        
dataframe.to_csv("dataset/raw_meta.csv")
```
